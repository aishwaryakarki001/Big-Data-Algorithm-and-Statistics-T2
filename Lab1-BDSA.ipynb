```python
import pandas as pd
from scipy import stats
```

```python
data= pd.read_csv('data.csv')
#print(data)
```

```python
print('Mean :',data.mean())
print('Median :',data.median())
print('Mode of A :', data['A'].mode().iloc[0])
print('Mode of B :', data['B'].mode().iloc[0])
print('SD :', data.std())
print('Variance :', data.var())
print('Trimmed Mean : A, B', stats.trim_mean(data,0.1))
```

```python
A_Q1= data['A'].quantile(0.25)
A_Q3= data['A'].quantile(0.75)

A_IQR = A_Q3 -A_Q1

print('Q1 OF A: ',A_Q1)
print('Q3 OF A: ',A_Q3)
print('IQR OF A: ',A_IQR)
```

```python
import matplotlib.pyplot as plt

# Assuming bins are already specified in the dataset
data.boxplot(column=['A', 'B'])
plt.title('Boxplot for Attribute A and B')
plt.show()

plt.hist(data['A'], bins=[0, 0.2, 0.4, 0.6, 0.8, 1], color='lightblue')
plt.title('Histogram for Attribute A')
plt.xlabel('Value')
plt.show()

plt.hist(data['B'], bins=[0, 5, 10, 15, 20], color='lightgreen')
plt.title('Histogram for Attribute B')
plt.xlabel('Value')
plt.show()
```

```python
plt.scatter(data['A'], data['B'], c='blue', marker='o')
plt.title('Scatter Plot of A to B')
plt.xlabel('Attribute A')
plt.ylabel('Attribute B')
plt.show()
```
